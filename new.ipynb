```python
# this process is done for each politician: isolate time, rename columns (trivial data cleansing), split tweet.
import pandas as pd
from datetime import datetime
bs = pd.read_json("data_master/resources/berniesanders.json",lines=True)
created_time = []
#step to get date from json
for index, row in bs.iterrows():
#     ts = time.strftime('%Y-%m-%d %H:%M:%S', time.strptime(row['created_at'],'%a %b %d %H:%M:%S +0000 %Y'))
    ts = datetime.strftime(row.created_at,'%Y-%m')
    created_time.append(ts)
#step to retrieve full tweets
bstweets = bs["full_text"]
t1_df= pd.DataFrame(bstweets)
#rename "full_text" to "tweets"
t1_df = t1_df.rename(columns={'full_text': 'tweet'})
# Insert a column with politician's affiliation
t1_df.insert(0, 'Political Affiliation', 'Democrats')
# Insert a column with politician's name
t1_df.insert(2, 'Politician', 'Bernie Sanders')
#the step to calculate total word count for each tweet
count = t1_df['tweet'].str.split().str.len()
t1_df['Word Count']=count
#create a column for date
t1_df['created_time'] =created_time
#create a column for number of  retweet
t1_df['number of retweets'] = bs["retweet_count"]
#create the column for number of likes of each tweets
t1_df['number of likes']=bs["favorite_count"]
t1_df.head()
```

```python
#exclude retweets
t1woRT_df =t1_df[~t1_df.tweet.str.contains("RT")]
t1woRT_df.head()
```

```python
#save to csv file
t1woRT_df.to_csv("data_master/resources/berniesanders_ft.csv", index=False)
```
