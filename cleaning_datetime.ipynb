```python
import pandas as pd
bs = pd.read_json("data_master/resources/berniesanders.json",lines=True)
```

```python
from datetime import datetime

create_time = []
for index, row in bs.iterrows():
#     ts = time.strftime('%Y-%m-%d %H:%M:%S', time.strptime(row['created_at'],'%a %b %d %H:%M:%S +0000 %Y'))
    ts = datetime.strftime(row.created_at,'%b %Y')
    create_time.append(ts)
create_time
```

```python
bstweets = bs["full_text"]
t1_df= pd.DataFrame(bstweets)
t1_df = t1_df.rename(columns={'full_text': 'tweet'})
t1_df.insert(0, 'Political Affiliation', 'Democrats')
t1_df.insert(2, 'Politician', 'Bernie Sanders')
count = t1_df['tweet'].str.split().str.len()
t1_df['Word Count']=count
t1_df['Create Time'] =create_time
t1_df.head()
```
